```python
y = data[:, 0]
```

```python
pixels = data[:, 1]
print (type(pixels))
print (len(pixels[0]))
print (pixels[10][10])
```

```python
X = np.zeros((pixels.shape[0], 48*48))

for ix in range(X.shape[0]):
    p = pixels[ix].split(' ')
    for iy in range(X.shape[1]):
        X[ix, iy] = int(p[iy])
```

```python
np.save('facial_data_X', X)
```
